```python
import matplotlib.pyplot as plt

for data in ["drag.dat", "drag_lower_area.dat"]:
    t = []
    v = []
    with open(data) as f:
        for l in f:
            tt, vv = l.split(' ')
            t.append(float(tt))
            v.append(float(vv))
        plt.plot(t, v)

plt.title('Time at constant power vs Velocity achieved riding a bike against air resistance')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()
```
